```python
# ============================================================
# Loan Prediction System using Machine Learning
#
# Objective:
# Build and compare multiple Machine Learning classification
# algorithms to predict whether a loan should be approved or rejected.
#
# Algorithms Used:
# 1. Logistic Regression
# 2. Decision Tree
# 3. Random Forest
# 4. K-Nearest Neighbors (KNN)
# 5. Support Vector Machine (SVM)
# 6. Naive Bayes
#
# Best Performing Model:
# Random Forest Classifier
# ============================================================

# ==========================================
# Step 1: Import Required Libraries
# ==========================================

import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
#import warnings
#warnings.filterwarnings('ignore')
```

```python
# ==========================================
# Step 2: Load the Dataset
# ==========================================

df = pd.read_csv('/content/drive/MyDrive/loan_data_new.csv')
df
```

```python
# =====================================================
# Step 4: Check Missing Values and Duplicate records
# =====================================================

df.isnull().sum()
df.duplicated().sum()
```

```python
df.dtypes
```

```python
# ==========================================
# Step 5: Encode Categorical Variables
# ==========================================

from sklearn.preprocessing import LabelEncoder

gender_encoder = LabelEncoder()
education_encoder = LabelEncoder()
home_encoder = LabelEncoder()
loan_intent_encoder = LabelEncoder()
previous_loan_encoder = LabelEncoder()

df['Gender'] = gender_encoder.fit_transform(df['Gender'])
df['Education'] = education_encoder.fit_transform(df['Education'])
df['Home Onwership'] = home_encoder.fit_transform(df['Home Onwership'])
df['Loan Intent'] = loan_intent_encoder.fit_transform(df['Loan Intent'])
df['Previous Loan'] = previous_loan_encoder.fit_transform(df['Previous Loan'])
df.head()
```

```python
# ==========================================
# Step 6: Feature Selection (X and y)
# ==========================================

x = df.drop('Loan Status', axis=1)
y = df['Loan Status']
```

```python
# ==========================================
# Step 7: Split Dataset into Training and Testing Sets
# ==========================================

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

len(x_train), len(x_test)
len(y_train), len(y_test)
```

```python
# ==========================================
# Step 8: Logistic Regression Model_training
# ==========================================

lr = LogisticRegression(random_state=42)
lr.fit(x_train, y_train)
```

```python
# Make Predictions
y_pred_lr = lr.predict(x_test)
y_pred_lr
```

```python
# Calculate Accuracy, Generate Confusion metrix and Classificate report
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
accuracy = accuracy_score(y_test, y_pred_lr)
print('accuracy is:', accuracy)

cm = confusion_matrix(y_test, y_pred_lr)
print(cm)

cr = classification_report(y_test, y_pred_lr)
print(cr)
```

```python
# ==========================================
# Step 9: Decision Tree Classifier Model_training
# ==========================================

dt = DecisionTreeClassifier(max_depth=5, min_samples_split=10)
dt.fit(x_train, y_train)
```

```python
# Make Predictions

y_pred = dt.predict(x_test)
y_pred
```

```python
# Calculate Accuracy, Generate Confusion metrix and Classificate report

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
accuracy = accuracy_score(y_test, y_pred)
print('accuracy is:', accuracy)

cm = confusion_matrix(y_test, y_pred)
print(cm)

cr = classification_report(y_test, y_pred)
print(cr)
```

```python
# ==========================================
# Step 10: Random Forest Classifier Model_training
# ==========================================

from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state= 42)
rf.fit(x_train, y_train)
```

```python
# Make Predictions

y_pred_rf = rf.predict(x_test)
y_pred_rf
```

```python
# Calculate Accuracy, Generate Confusion metrix and Classificate report

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

accuracy_rf = accuracy_score(y_test, y_pred_rf)
print('Random_Forest_accuracy is:', accuracy_rf)

cm_rf = confusion_matrix(y_test, y_pred_rf)
print(cm_rf)

cr_rf = classification_report(y_test, y_pred_rf)
print(cr_rf)
```

```python
# ==========================================
# Step 11: K-Nearest Neighbors (KNN) Model_training
# ==========================================

from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(x_train, y_train)
```

```python
# Make Predictions

y_pred_knn = knn.predict(x_test)
y_pred_knn
```

```python
# Calculate Accuracy, Generate Confusion metrix and Classificate report

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

accuracy_knn = accuracy_score(y_test, y_pred_knn)
print('KNN_accuracy is:', accuracy_knn)

cm_knn = confusion_matrix(y_test, y_pred_knn)
print(cm_knn)

cr_knn = classification_report(y_test, y_pred_knn)
print(cr_knn)
```

```python
# ==========================================
# Step 12: Support Vector Machine (SVM) Model_training
# ==========================================

from sklearn.svm import SVC
svm = SVC(random_state = 42)
svm.fit(x_train, y_train)
```

```python
# Make Predictions

y_pred_svm = svm.predict(x_test)
y_pred_svm
```

```python
# Calculate Accuracy, Generate Confusion metrix and Classificate report

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
accuracy_svm = accuracy_score(y_test, y_pred_svm)
print('SVM_accuracy is:', accuracy_svm)

cm_svm = confusion_matrix(y_test, y_pred_svm)
print(cm_svm)

cr_svm = classification_report(y_test, y_pred_svm)
print(cr_svm)
```

```python
# ==========================================
# Step 13: Naive Bayes Classifier Model_training
# ==========================================

from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(x_train, y_train)
```

```python
# Make Predictions

y_pred_nb = nb.predict(x_test)
y_pred_nb
```

```python
# Calculate Accuracy, Generate Confusion metrix and Classificate report

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

accuracy_nb = accuracy_score(y_test, y_pred_nb)
print('NB_accuracy is:', accuracy_nb)

cm_nb = confusion_matrix(y_test, y_pred_nb)
print(cm_nb)

cr_nb = classification_report(y_test, y_pred_nb)
print(cr_nb)
```

```python
# ==========================================
# Step 14: Final Model Selection
# ==========================================

# Random Forest achieved the best performance among all models.
# Therefore, it was selected as the final model for deployment.
```
